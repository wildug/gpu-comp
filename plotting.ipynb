```python
import matplotlib.pyplot as plt

# Sample data
algorithms = ['cuBLAS_float', 'cuBLAS_int', 'MATVANS']
problem_sizes = [512, 1024, 2048, 4096, 6144]


# Format: [algo][problem_size]
with_memcpy = [
    [2.8, 8.4, 29, 123, 302],     # cuBLAS_float
    [1.6, 4.6, 15.7, 70, 171],     # cuBLAS_int
    [4, 10, 23, 59, 108],     # MULTANS
]
without_memcpy = [
    [0.5 , 0.6,  1,    3, 6],   # cuBLAS_float
    [0.87, 2.25, 7.9, 41, 108],     # cuBLAS_int
    [3   , 8   , 15, 31,   47],   # MULTANS
]


# Plotting
plt.figure(figsize=(10, 6))
markers = ['o', 's', 'D']  # Circle, square, diamond
colors = [
    "#E69F00",  # orange
    "#CC79A7",  # purple
    "#56B4E9",  # blue
    "#009E73",  # green
    "#F0E442",  # yellow
    "#0072B2",  # dark blue
    "#D55E00",  # red-orange
]

for i, algo in enumerate(algorithms):
    plt.plot(problem_sizes, without_memcpy[i],
             label=f'{algo} (without memcpy)', linestyle='--', marker="o",markerfacecolor='none', color=colors[i], alpha=0.7)
    plt.plot(problem_sizes, with_memcpy[i],
             label=f'{algo} (with memcpy)', linestyle='-', marker="o", color=colors[i])

# Labels and legend
plt.xlabel('Row and columns size of $W_k$')
plt.ylabel('Time (ms)')
plt.xscale("log", base=2)
# plt.title('Latency as a function of matrix size with and without memcpy')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.savefig("latencies.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
